```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset.House.txt')
df.head()
```

```python
X = df.drop(['Price'],axis=1)
X = pd.get_dummies(X, columns=['Location'], drop_first=True)
X.to_numpy()
```

```python
inputs = df['Price'].to_numpy()
Y = inputs
Y
```

```python
obj_col = df.select_dtypes(include='object').columns.tolist()
print("categorical variables are: ",len(obj_col))
print("categorical columns: ",obj_col)

num_col = df.select_dtypes(include='number').columns.tolist()
print("numerical variables: ",len(num_col))
print("numerical columns: ",num_col)
```

```python
number_data = df.select_dtypes(include='number')

import seaborn as sns
plt.figure(figsize=(6,3))
sns.heatmap(number_data.corr(), annot=True, cmap='BrBG',linewidths=2,fmt='.2f')
plt.title("House Price Prediction For HEATmap")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X,Y)
ycap = model.predict(X)
acc = r2_score(Y,ycap)
print("Linear Regression for Prediction:", acc)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X,Y)
ycap = model.predict(X)
acc = r2_score(Y,ycap)
acc
```
